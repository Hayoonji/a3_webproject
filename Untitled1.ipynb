## 가설3. 키움(전 넥슨) 히어로즈의 2016년 경기장 이전으로 인한 타격 지표의 변화

기존의 목동 구장에서 고척돔으로 옮기면서 달라진 구장에 관련하여 지표가 어떻게 달라졌는지 알아보자.

- 타율, 출루, 장타

```python
# 살펴볼 타격 지표인 타율 출루 장타 삼진 병살을 추출
baseball = baseball[['이름', '연도','타율', '출루', '장타']]
baseball
```

```python
baseball['이름'].unique()
```

```python
nexen = baseball[(baseball.이름=='넥센') | (baseball.이름 =='키움')]
nexen
```

우선 연도별로 알아보기 위해 연도에 따라 그룹을 나눈 뒤 평균을 구한다.

```python
nexen2 = nexen.groupby(['연도'], as_index=False).mean()
nexen2
```

subplot을 통해 연도별 타율, 출루율, 장타율의 평균을 선 그래프로 그려 변화를 비교한다

```python
# 그래프 크기 설정
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# 첫 번째 서브플롯에 선 그래프 그리기
sns.lineplot(x=nexen2['연도'], y=nexen2['타율'], ax=ax[0], color='#fc0000')
ax[0].set_title('타율')

# 두 번째 서브플롯에 선 그래프 그리기
sns.lineplot(x=nexen2['연도'], y=nexen2['출루'], ax=ax[2], color='#0000ff')
ax[1].set_title('출루')

# 세 번째 서브플롯에 선 그래프 그리기
sns.lineplot(x=nexen2['연도'], y=nexen2['장타'], ax=ax[1], color='#f2da00')
ax[2].set_title('장타')


# X 축 눈금 및 라벨 설정
for axis in ax:
    axis.set_xticks(nexen2['연도'])
    axis.set_xticklabels(nexen2['연도'])

# 그래프 간 간격 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

반발계수 변화 후와 전의 타격지표를 막대 그래프로 비교한다.

```python
nexen2

nexen2['구장 변화'] = nexen2['연도'].apply(lambda x : '구장 변화 전' if x <= 15 else "구장 변화 후" )

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

sns.barplot(x = '구장 변화',y = "타율", data = nexen2, ax = ax[0], palette="Set1")
ax[0].set_ylim(0.200, 0.500)
ax[0].set_title('타율')

sns.barplot(x = '구장 변화',y = "출루", data = nexen2, ax = ax[1], palette="Set1")
ax[1].set_ylim(0.200, 0.500)
ax[1].set_title('출루')

sns.barplot(x = '구장 변화',y = "장타", data = nexen2, ax = ax[2], palette="Set1")
ax[2].set_ylim(0.200, 0.500)
ax[2].set_title('장타')
```
